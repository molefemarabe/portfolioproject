```python
from google.colab import files
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd

uploaded = files.upload()

df = pd.read_csv(next(iter(uploaded)), encoding = 'cp1252') #Store data in a DataFrame after upload
df.head(10) #preview first 10 rows of the data

```

```python
duplicates = df.duplicated().sum()
print(duplicates)
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df.head(3)
```

```python
df['Period'] = df['Order Date'].dt.to_period('M')
```

```python
df[['Order Date','Ship Date', 'Period']].head(3)
```

```python
df['Profit Margin'] = df['Profit'] / df['Sales']
```

```python
df.head(2)
```

```python
avg_margin_by_period = df.groupby('Period')['Profit Margin'].mean()
```

```python
avg_margin_by_period.plot(kind='line', title='Average Profit Margin by Period')
plt.ylabel('Avarage Profit Margin')
plt.show()
```

```python
df.groupby('Category')['Sales'].sum().sort_values(ascending = True)
```

```python
monthly_sales = df.groupby('Period')['Sales'].sum()
```

```python
monthly_sales.head(4)
```

```python
monthly_sales.plot(kind='line', figsize=(10,5), title='Monthly Sales Trend')
plt.ylabel('Total Sales')
```
